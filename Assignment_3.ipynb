EXERCISE 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Loop over different values of n
for n in [500, 1000, 2000, 5000, 10000, 15000, 20000, 50000, 100000]:
    # Step 2: Simulate throwing two dice n times and compute the sum
    dice_1 = np.random.randint(1, 7, n)
    dice_2 = np.random.randint(1, 7, n)
    dice_sum = dice_1 + dice_2

    # Step 3: Compute the frequencies using numpy's histogram function
    h, bins = np.histogram(dice_sum, bins=np.arange(2, 14))

    # Step 4: Plot the histogram using matplotlib's bar function
    plt.bar(bins[:-1], h / n, width=1.0)
    plt.title(f'Histogram of Dice Sum (n={n})')
    plt.xlabel('Sum of Dice')
    plt.ylabel('Frequency')
    plt.show()

    # Step 5: Observations
    print(f"Observations for n = {n}:")
    print("- As n increases, the shape of the histogram tends to resemble a normal distribution.")
    print("- The frequencies of sums around the mean value are higher compared to extreme sums.")
    print("- There is a tendency for the histogram to become smoother as n increases.")

    # Step 6: Relation to "regression to the mean"
    print("Relation to 'regression to the mean':")
    print("- 'Regression to the mean' refers to the tendency for extreme outcomes to move closer to the average over repeated trials.")
    print("- In this context, as n increases, the distribution of the dice sum tends to converge towards the expected mean of 7.")
    print("- Extreme outcomes, such as sums of 2 or 12, become less frequent as n increases, exhibiting 'regression to the mean'.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Load the data
data = pd.read_csv('/content/weight-height 1.csv')
# Step 2: Create a scatter plot to inspect the dependence between height and weight
plt.figure(figsize=(8, 6))
plt.scatter(data['Height'], data['Weight'], color='blue')
plt.title('Scatter Plot of Height vs Weight')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (lbs)')
plt.grid(True)
plt.show()
# Step 3: Perform linear regression
X = data['Height'].values.reshape(-1, 1)
y = data['Weight'].values

model = LinearRegression()
model.fit(X, y)

# Step 4: Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red', linewidth=2)
plt.title('Linear Regression: Height vs Weight')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (lbs)')
plt.grid(True)
plt.show()

# Step 5: Compute RMSE and R2 value
y_pred = model.predict(X)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2) value:", r2)
```
